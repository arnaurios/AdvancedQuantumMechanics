# Scattering in spherical coordinates
We consider the scattering of two particles interacting through a barrier in the $L=0$ (or s-) wave. 

## Repulsive well
### Scattering below the barrier
The case of a repulsive ($V_0>0$) square well potential, 
$$
V(r) =\begin{cases}
V_0, \, r \leq R , \\
0, \, r > R \, .
\end{cases}
$$
The potential is schematically depicted in the Figure below:
<img src="img/well_repulsive.png" alt="drawing" width="300"/>
We consider an initial situation in which the energy of the projectile is lower than the height of the barrier, $E<V_0$. This corresponds to the gray line above. We will consider the situation of scattering of energies above the barrier, $E>V_0$, later on.

The Schr√∂dinger equation for the radial component of the wavefunction
is simplified if we consider the radial wavefunction $R_l(k,r)=\frac{u_l(k,r)}{r}$. With this change, the equation
becomes:
$$
- \frac{\hbar^2}{2 m} \frac{1}{r} \frac{d^2}{dr^2} u_l(k,r)
+ \frac{\hbar^2}{2 m} \frac{l(l+1)}{r^3} u_{l}( k, r ) + V(r) \frac{u_l(k,r)}{r} = E \frac{u_l(k,r)}{r} \, .
$$
Multiplying by $r$ and $-2m/\hbar^2$, we get:
$$
 \frac{d^2}{dr^2} u_{l}( k, r ) - \frac{l(l+1)}{r^2} u_l(k,r) - \frac{2mV(r)}{\hbar^2}  u_l(k,r)  = -\frac{2mE}{\hbar^2} u_l(k,r) \, .
$$
We consider the $S-$wave scattering case, for which $l=0$. In this case, the centrifugal term drops 
and the equation is a second-order homogeneous differential equation:
$$
 \frac{d^2}{dr^2} u_0( k, r ) + \frac{2m [E - V(r)]}{\hbar^2}  u_0(k,r)  =0 \, .
$$
For the repulsive square well, we have two regions of space with different equations depending on whether $r<R$ or 
$r>R$,
$$
 \frac{d^2 u_0}{dr^2} + \frac{2m [E- V_0]}{\hbar^2}  u_0  =0 \, , \quad  r \leq R , \\
 \frac{d^2 u_0}{dr^2} + \frac{2m  E}{\hbar^2}  u_0  =0 \, , \quad  r > R .
$$
Introducing the wave-vector $k^2=\frac{2m  E}{\hbar^2} $, the second equation is just an
harmonic equation. The first equation is also harmonic similar, but with a slightly
different wave-vector, $\bar k^2= k^2 - k_0^2$, where 
we define the momentum scale associated to the square well as 
$k_0=\sqrt{ \frac{ 2m V_0}{\hbar^2}}$. With this, the two equations are
$$
 \frac{d^2 u_0}{dr^2} + \bar k^2  u_0  =0 \, , \quad  r \leq R , \label{eq:well1} \\
 \frac{d^2 u_0}{dr^2} + k^2  u_0  =0 \, , \quad  r > R .
$$
The solution to the first equation is different depending on whether $\bar k^2$ is 
positive or negative. 
We start with the negative case, which corresponds to $E<V_0$, and hence $k^2 < k_0^2$. 
Physically, this represents scattering at energies
below the barrier of the well (gray line in the figure). Mathematically, this means that $\bar k$ is a pure imaginary
number, $\bar k = i \kappa$,
with $\kappa=\sqrt{k_0^2 - k^2}>0$. 

The solution of the two equations in the two spatial regions are essentially exponentials of 
imaginary or real numbers times $r$. The solutions in the interior region, $r < R$, should be
regular at the origin, so $u_0( r \to 0) \to 0$. In contrast, in the exterior region, $r > R$, the
waves are asymptotically free but modified by the phase shift $u_0( r \to \infty) \to \sin( kr + \delta_0)$.
With this, we can immediately guess the following two solutions:
$$
 \frac{d^2 u_0}{dr^2}  + \bar k^2  u_0  =0 \, ,  r \leq R 
 \Rightarrow u_0(r)=A \sinh \kappa r \\
 \frac{d^2 u_0}{dr^2}  + k^2  u_0 =0 \, , r > R 
  \Rightarrow u_0(r)=B \sin \left( k r + \delta_0 \right).
$$
To determine $\delta_0$, we look at the logarithmic derivative of the 
wave-functions at values $R_-=R-\epsilon$, right below $R$, and at 
$R_+=R+\epsilon$, right above $R$ with $\epsilon \to 0$. 
A physically valid matching can only be achieved if the
wavefunctions fulfil $u_0(R_-)=u_0(R_+)$ and if their derivatives fulfill 
$\left. \frac{d u_0}{dr} \right|_{r=R_-}=\left.  \frac{d u_0}{dr} \right|_{r=R_+}$. In consequence, the logarithmic derivatives must be such that:
$$
\frac{1}{u_0}\left.  \frac{d u_0}{dr} \right|_{R_-} =  \frac{1}{u_0} \left. \frac{d u_0}{dr} \right|_{R_+}. 
$$
Taking the logarithmic derivatives for the solutions, we find:
$$
\frac{1}{A \sinh \kappa R} A \kappa \cosh \kappa R  = 
\frac{1}{B \sin \left( k R + \delta_0 \right)} B \sin \left( k R + \delta_0 \right) \Rightarrow
\frac{1}{\tanh \kappa R} = 
\frac{1}{\tan  \left( k R + \delta_0 \right)} 
$$
This is an equation for the phaseshift, which can be isolated to find:
\begin{align}
\delta_0 = \arctan \left[ \frac{k}{\kappa} \tanh \left( \kappa R \right) \right] - k R \nonumber = \arctan  \left[ \frac{k}{\sqrt{ k_0^2-k^2} } \tanh \left( \sqrt{ k_0^2-k^2} R \right) \right] - k R \, ,
\end{align}

### Code
We start by choosing a value of $R$ and the height of the potential (or, equivalently, a value of the momentum $k_0=\sqrt{ \frac{ 2m V_0}{\hbar^2}}$). We then set up functions for the phaseshifts below and above the barrier, and compute them for a serues of energies,

```python
# coding: utf-8
#########################################################################################
# THIS PYTHON CODE GIVES THE PHASE SHIFTS FOR SCATTERING OFF A REPULSIVE BARRIER
#########################################################################################
import math
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle


zi=complex(0,1);
pi=math.pi
epsilon=0.001

def phaseshift_barrier_below(k,k0,R) :
    kappa=np.sqrt(np.power(k0,2) - np.power(k,2) )
    ph=( np.arctan2( k*np.tanh(kappa*R),kappa ) - k*R )/pi
    return ph

# THE RADIUS OF THE SCATTERING BARRIER
R=1;

# THE MOMENTUM OF THE SCATTERING BARRIER
k0=1;

# THE ENERGY WE ARE LOOKING AT
k_momentum=0.5;
rmax=pi/k_momentum*3
rr=np.linspace(epsilon,rmax,200)

fig, ax = plt.subplots(figsize=(8,6), ncols=1, nrows=1)

lab="L=0"
ps=phaseshift_barrier_below(k_momentum,k0,R)*pi
kappa=np.sqrt(np.power(k0,2) - np.power(k_momentum,2) )

wf0=np.sinh(kappa*rr)
ab=np.sinh(kappa*R)/np.sin(k_momentum*R+ps)
wf1=ab*np.sin(k_momentum*rr+ps)

c_full = plt.cm.Blues(np.linspace(0.3,1,2))   
#ax.plot(rr[np.where(rr>R)],wf1[np.where(rr>R)],label="$R_l$",color=c_full[1])
ax.plot(rr[np.where(rr>R)],wf1[np.where(rr>R)],label="Exterior sol",color=c_full[1])
ax.plot(rr[np.where(rr<R)],wf0[np.where(rr<R)],label="Interior sol",color=c_full[1],linestyle="dashed")

ax.legend()
ymax=(np.floor( np.amax(wf1)*10 ) + 2)/10
print(ymax)
rect = Rectangle((0, -ymax), R, 2*ymax, linewidth=1, facecolor='#D9DAD8')
ax.add_patch(rect)
ax.set_ylim([-ymax,ymax])
ax.set_xlim([0,rmax])
ax.text(0.9, 0.85, lab,transform=ax.transAxes,fontsize=15)
    
ax.set_ylabel(r"Wave function,  $u_0(kr)$",fontsize=18)
ax.set_title("Below barrier scattering, $R=$"+str(R)+", k_0="+str(k0) + ", k="+str(k_momentum),fontsize=18)
```

### Scattering above the barrier
For $k>k_0$, the solution is mathematically different. The momentum $\bar k=\sqrt{k^2-k_0^2}$ is not imaginary anymore, but a real number. As a consequence,
the solution to the first equation (in the interior region) is different and becomes a $\sin(x)$ instead of a 
$\sinh(x)$:
 \begin{align}
 \bar k^2 &= k^2 - k_0^2 \\
 \frac{d^2 u_0}{dr^2}  & + \bar k^2  u_0  =0 \, ,  r \leq R 
 \Rightarrow u_0(r)=A \sin \bar k r \\
 \frac{d^2 u_0}{dr^2}  & + k^2  u_0 =0 \, , r > R 
  \Rightarrow u_0(r)=B \sin \left( k r + \delta_0 \right). 
\end{align}
We can still find the phase-shift $\delta_0$ by imposing the continuity of logarithmic derivativeswhich in this case leads to 
\begin{align}
\frac{\bar k}{\tan \bar k R} &= 
\frac{k}{\tan  \left( k R + \delta_0 \right)} 
\label{eq:logder1}
\end{align}
Isolating the phaseshift, we find:
\begin{align}
\delta_0 
&= \arctan  \left[ \frac{k}{\sqrt{ k^2-k^2} } \tan \left( \sqrt{ k^2 - k_0^2} R \right) \right] - k R \, ,
\label{eq:ps1}
\end{align}
Importantly, this expression matches perfectly well with the case of scattering below the barrier in 
the $k \to k_0$ limit, even though the arguments of the $\arctan$ function diverge in the two expressions.


```python
# coding: utf-8
#########################################################################################
# THIS PYTHON CODE GIVES THE PHASE SHIFTS FOR SCATTERING OFF A REPULSIVE BARRIER
#########################################################################################
import math
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle


zi=complex(0,1);
pi=math.pi
epsilon=0.001

def phaseshift_barrier_above(k,k0,R) :
    kappa=np.sqrt(np.power(k,2) - np.power(k0,2) )
    ph0=( np.arctan2( k*np.tan(kappa*R),kappa ) - k*R )%pi
    ph=ph0/pi-1
    return ph

# THE RADIUS OF THE SCATTERING BARRIER
R=1;
# THE MOMENTUM OF THE SCATTERING BARRIER
k0=1;
# THE ENERGY WE ARE LOOKING AT
k_momentum=2;
rmax=pi/k_momentum*3
rr=np.linspace(epsilon,rmax,200)

fig, ax = plt.subplots(figsize=(8,6), ncols=1, nrows=1)

lab="L=0"
ps=phaseshift_barrier_above(k_momentum,k0,R)*pi
kappa=np.sqrt(np.power(k_momentum,2) - np.power(k0,2) )

wf0=np.sin(kappa*rr)
ab=np.sin(kappa*R)/np.sin(k_momentum*R+ps)
wf1=ab*np.sin(k_momentum*rr+ps)

c_full = plt.cm.Blues(np.linspace(0.3,1,2))   
#ax.plot(rr[np.where(rr>R)],wf1[np.where(rr>R)],label="$R_l$",color=c_full[1])
ax.plot(rr[np.where(rr>R)],wf1[np.where(rr>R)],label="Exterior sol",color=c_full[1])
ax.plot(rr[np.where(rr<R)],wf0[np.where(rr<R)],label="Interior sol",color=c_full[1],linestyle="dashed")

ax.legend()
ymax=(np.floor( np.amax(wf1)*10 ) + 2)/10
print(ymax)
rect = Rectangle((0, -ymax), R, 2*ymax, linewidth=1, facecolor='#D9DAD8')
ax.add_patch(rect)
ax.set_ylim([-ymax,ymax])
ax.set_xlim([0,rmax])
ax.text(0.9, 0.85, lab,transform=ax.transAxes,fontsize=15)
    
ax.set_ylabel(r"Wave function,  $u_0(r)$",fontsize=18)
ax.set_title("Below barrier scattering, $R=$"+str(R)+", k_0="+str(k0) + ", k="+str(k_momentum),fontsize=18)
```

### Phase shift


```python
# coding: utf-8
#########################################################################################
# THIS PYTHON CODE GIVES THE PHASE SHIFTS FOR SCATTERING OFF A REPULSIVE BARRIER
#########################################################################################
import math
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
from scipy.special import spherical_jn,spherical_yn

zi=complex(0,1);
pi=math.pi
epsilon=0.001

# THE RADIUS OF THE SCATTERING BARRIER
R=1;

# THE MOMENTUM OF THE SCATTERING BARRIER
k0=1;


def phaseshift_barrier_above(k,k0,R) :
    kappa=np.sqrt(np.power(k,2) - np.power(k0,2) )
    ph0=( np.arctan2( k*np.tan(kappa*R),kappa ) - k*R )%pi
    ph=ph0/pi-1
    return ph


# COMPUTE PHASE-SHIFT BELOW THE BARRIER
Nk_below=40
kk1=np.linspace(0,k0-epsilon,Nk_below)
ps1=phaseshift_barrier_below(kk1,k0,R)

# COMPUTE PHASE-SHIFT ABOVE THE BARRIER
Nk_above=40
kk2=np.linspace(k0+epsilon,6,Nk_below)
ps2=phaseshift_barrier_above(kk2,k0,R)

# RESONANCE CONDITION
kres1=np.sqrt( np.power(k0,2) + (pi/(2*R)))

fig, (ax1) = plt.subplots(figsize=(5,4), ncols=1, nrows=1)
ax1.plot(kk1,ps1,color="#045a8d")
ax1.plot(k0,phaseshift_barrier_below(k0-0.001,k0,R),'o',color="#74a9cf")

ax1.plot(kk2,ps2,"--",color="#2b8cbe")
ax1.plot(kres1,phaseshift_barrier_above(kres1,k0,R),'s',color="#74a9cf")

ax1.set_ylabel(r"Phase-shift, $\delta/\pi$",fontsize=18)
ax1.set_title(r"Barrier momentum, $k_0=$"+str(k0)+"\n" + "Barrier radius, $R=$"+str(R),fontsize=18)
```

### Scattering length
The phaseshift as a function of $k$ is shown in the Figure above for 
an arbitrary value of $k_0=0$. The solid line corresponds to the case
$k<k_0$ that we have just discussed. The expression is valid
from $k=0$ to $k=k_0$ (circle). 
The behaviour at low values of momenta, $k \ll k_0$, appears to be linear.
This can be justified from a low-momentum expansion in the phase-shift equation, which indicates 
\begin{align}
\delta_0 & \approx \arctan  \left[ \frac{k}{k_0 } \tanh k_0 R \right] - k R 
\approx \left[ \frac{k}{k_0 } \tanh k_0 R \right] - k R \\
&=- k R \left[1 - \frac{ \tanh k_0 R}{k_0 R} \right] \equiv - k a_s 
\label{eq:ps0}
\end{align}
The (negative of the) constant of proportionality between $\delta_0$ and $k$ at low energies 
is the so-called **scattering length**, $a_s$. This only depends on the properties of 
the potential and for a repulsive square well it is given by the expression
\begin{align}
a_s = R \left[1 - \frac{ \tanh k_0 R}{k_0 R} \right] 
\label{eq:as_barrier}
\end{align}
Because $\tanh x < 1$, the scattering length of a repulsive potential well is always
positive, $a_s>0$. 
How can we relate these low-momentum properties to the cross section? The $l=0$ component of 
the total cross section is
\begin{align}
\sigma_0 = \frac{4 \pi}{k^2} \sin^2 \delta_0 \xrightarrow[ k \to 0 ]{} \frac{4 \pi}{k^2} \sin^2 (- k a_s) \approx 
4 \pi a_s^2
\end{align}
This corresponds to the cross section of a sphere of effective radius $a_s$. 
At low energies, the scattering wave ``sees" a portion of the potential characterised by $a_s$. 


### Divergences
This is not the only divergence ``hidden" in the phaseshifts. Whenever the argument of the $\tan$ function
in the first term becomes a multiple of $\pi/2$, the function itself will diverge. This happens
when the condition
\begin{align*}
\sqrt{ k_n^2 - k_0^2} R=\frac{\pi}{2} n \Rightarrow
k_n=\sqrt{k_0^2 + \left(  \frac{\pi}{2} \frac{n}{R} \right)^2}
\end{align*}
The divergence is not apparent 
in the phase shift because the $\arctan$ function itself does not diverge for infinite arguments. 
The position of $k_1$ is marked with a square in the Figure.


## Attractive well in s wave
We now consider a similar, yet physically different, problem. This is now an attractive square well,
\begin{align}
V(r) =\begin{cases}
-V_0, \, r \leq R , \\
0, \, r > R \, ,
\end{cases}
\end{align}
The potential is schematically depicted in the Figure below:
<img src="img/well_attractive.png" alt="drawing" width="300"/>

Mathematically, the problem is very similar to the cases above. The differential equations in the two regions
of space are now:
 \begin{align}
 \frac{d^2 u_0}{dr^2}  & + \bar{ \bar{ k}}^2  u_0  =0 \, ,  r \leq R 
 \Rightarrow u_0(r)=A \sin \bar{ \bar{ k}} r \\
 \frac{d^2 u_0}{dr^2}  & + k^2  u_0 =0 \, , r > R 
  \Rightarrow u_0(r)=B \sin \left( k r + \delta_0 \right). 
\end{align}
where now the relevant momentum is real
 \begin{align}
\bar{ \bar{ k}} = \sqrt{ k^2 + k_0^2 } \, .
\end{align}
The phase-shift $\delta_0$ is found again by matching logarithmic derivatives, 
 so that
\begin{align*}
\frac{{\bar {\bar k}} }{\tan {\bar {\bar k}} R} &= 
\frac{k}{\tan  \left( k R + \delta_0 \right)} \, ,
\end{align*}
which yields
\begin{align}
\delta_0 %&= \arctan \left[ \frac{k}{\bar k} \tan \left( \bar k R \right) \right] - k R \\
&= \arctan  \left[ \frac{k}{\sqrt{ k_0^2+k^2} } \tan \left( \sqrt{ k_0^2+k^2} R \right) \right] - k R \, . 
\label{eq:phase_attractive}
\end{align}


```python
def phaseshift_well(k,k0,R) :
    kappa=np.sqrt(np.power(k0,2) + np.power(k,2) )
    ph0=( np.arctan2( k*np.tan(kappa*R),kappa ) - k*R )%pi
    ph=ph0/pi
    return ph

# THE RADIUS OF THE SCATTERING BARRIER
R=1;
k0=1;
kcrit=pi/2/R

fig, axs = plt.subplots(figsize=(5,4), ncols=1, nrows=1, )

# THE MOMENTUM OF THE SCATTERING BARRIER

# COMPUTE PHASE-SHIFT BELOW THE BARRIER
Nk_below=100
kk1=np.linspace(epsilon,k0-epsilon,Nk_below)
ps1=phaseshift_well(kk1,k0,R)

# COMPUTE PHASE-SHIFT ABOVE THE BARRIER
Nk_above=100
kk2=np.linspace(k0+0.001,6,Nk_below)
ps2=phaseshift_well(kk2,k0,R)


plt.plot(kk1,ps1,color="#045a8d")
plt.plot(kcrit,phaseshift_well(kcrit,k0,R),'s',color="#74a9cf")

plt.plot(kk2,ps2,"--",color="#2b8cbe")

plt.xlabel("Momentum, k",fontsize=18)
plt.ylabel(r"Phase-shift, $\delta/\pi$",fontsize=18)
#    plt.title(r"Well momentum, $k_0=$"+"{:.3f}".format(k0)+"\n" + "Well radius, $R=$"+str(R),fontsize=18)
plt.title("Well radius, $R=$"+str(R),fontsize=18)
```

Clearly, at low energies the phase shift is linear in $k$. We can find the constant of proportionality
by looking at the phase shift at low momentum,
\begin{align}
\delta_0 \approx \frac{k}{ k_0 } \tan \left( k_0 R \right) - k R =
- k \underbrace{  \left[ R - \frac{\tan k_0 R}{k_0} \right] }_{=a_s} \, ,
\end{align}
which indicates that the scattering length in this case is
\begin{align}
a_s = R \left[ 1 - \frac{\tan k_0 R}{k_0R} \right] \, .
\label{eq:as_attractive}
\end{align}
Because $\tan x > x$, the scattering length of this potential is **negative**, in  contrast to the
repulsive case. The phase shift thus grows linearly with $k$ at low energies.
